```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datosT= pd.read_csv('DatosTrafico2', index_col=0)
```

```python
datosT1=datosT.copy()
datosT1.drop([0], axis=0, inplace = True)
datosT1=datosT1.astype(float)
```

```python
import networkx as nx
from gglasso.problem import glasso_problem
from gglasso.helper.utils import sparsity, zero_replacement, normalize, log_transform
from gglasso.helper.basic_linalg import scale_array_by_diagonal
```

```python
def mean_norm(datosT1):
    return datosT1.apply(lambda x: (x-x.mean()), axis=0)
```

```python
datosT3 = mean_norm(datosT1) #Son los datos estandarizados con media 0 CREO (ver lo de x_i~N(0,sigma))
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
from sklearn import preprocessing
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
valoreslanda=np.logspace(-0.7, -4, 34)
erroresmse=np.zeros((34,500))
erroresmae=np.zeros((34,500))
errormseminlanda=np.linspace(0,0,34)
errormaeminlanda=np.linspace(0,0,34)
cantidaddeneuronasmse=np.linspace(0,0,34)
cantidaddeneuronasmae=np.linspace(0,0,34)
(p,N) = (datosT3.T).shape
S0 = np.cov(datosT3.T, bias = True)
lambda1_rango = valoreslanda
sel_caract=np.linspace(0,0,34)
for j in range(0,34):
    valor_lambda= lambda1_rango[j]
    P = glasso_problem(S0, N, reg_params = {'lambda1': valor_lambda}, latent = False, do_scaling = False)
    P.solve(solver_params={'max_iter':100000})
    sol = P.solution.precision_
    P.solution.calc_adjacency(t = 1e-4)
    VS=pd.DataFrame() #DataFrame vacio
    sum=0
    for i in range(294):
        if abs(sol[i,293])>=0.0008:
           sum += 1
           VS=pd.concat([VS,datosT3[str(i)]], axis=1)
    sel_caract[j]=sum
    y=VS[str(293)]
    X=VS
    X.drop([str(293)], axis=1, inplace=True)
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
    scaler = StandardScaler()
    X_trainscaled=scaler.fit_transform(X_train)
    X_testscaled=scaler.transform(X_test)
    nn=np.linspace(1, 500, 500)
    errores1=np.linspace(0,0,500)
    errores2=np.linspace(0,0,500)
    for m in range(1,501):
        mlp=MLPRegressor(hidden_layer_sizes=(m,),  random_state=1, solver='adam', activation="relu",max_iter=3000000, early_stopping=True).fit(X_trainscaled, y_train.values)
        y_pred=mlp.predict(X_testscaled)
        mse=mean_squared_error(y_pred,y_test)
        mae=mean_absolute_error(y_pred,y_test)
        errores1[m-1]=mse
        errores2[m-1]=mae
    erroresmse[j]=errores1
    erroresmae[j]=errores2
    lugardelerror1min=np.argmin(errores1)
    lugardelerror2min=np.argmin(errores2)
    errormseminlanda[j]=errores1[lugardelerror1min]
    errormaeminlanda[j]=errores2[lugardelerror2min]
    cantidaddeneuronasmse[j]=lugardelerror1min + 1
    cantidaddeneuronasmae[j]=lugardelerror2min + 1
```
